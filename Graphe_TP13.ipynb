```r
library(ggplot2)
library(dplyr)

# Charger les données diamonds
data("diamonds")

# Filtrer les données pour rendre le graphique plus lisible (optionnel)
diamonds_filtered <- diamonds %>% filter(price < 15000)

# Créer le graphique de densité
p <- ggplot(diamonds_filtered, aes(x = price, fill = clarity)) +
  geom_density(alpha = 0.5) +
  facet_wrap(~ cut) +
  scale_x_log10() +
  theme_minimal() +
  labs(title = "Distribution des prix des diamants par qualité de coupe et clarté",
       x = "Prix (log10)",
       y = "Densité",
       fill = "Clarté")

# Afficher le graphique
print(p)
```

```r
library(ggplot2)

# Charger les données diamonds
data("diamonds")
# Filtrer les données directement avec ggplot2 (optionnel)
diamonds_filtered <- subset(diamonds, price < 15000)

# Créer le graphique de densité
p <- ggplot(diamonds_filtered, aes(x = price, fill = clarity)) +
  geom_density(alpha = 0.5) +
  facet_wrap(~ cut) +
  scale_x_log10() +
  theme_minimal() +
  labs(title = "Distribution des prix des diamants par qualité de coupe et clarté",
       x = "Prix (log10)",
       y = "Densité",
       fill = "Clarté")

# Afficher le graphique
print(p)
```
